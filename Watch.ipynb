```python
# Example of dealing with outliers using Z-score
for col in nr:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 2.5 * IQR
    upper_bound = Q3 + 2.5 * IQR
    outliers=df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(col,'Q1:',Q1,'Q3:' ,Q3, 'ITQ:',IQR, 'Lower bound:',lower_bound,'Upper bound', upper_bound, '\n','Outliers:',outliers.index,'\n')

    sns.scatterplot(data=df,y=col,x='Price (USD)',label='Normal data')
    sns.scatterplot(data=outliers,y=col,color='red',x='Price (USD)',label='Outliers')
    plt.show()
    
    df=df[~((df[col] < lower_bound) | (df[col] > upper_bound))]
```
